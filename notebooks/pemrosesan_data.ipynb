```python
import pandas as pd
```

#### di sini langkah awal kita adalah dengan membuat penampungan dari semua file dataset kita

```python
file_saham={
    'bbca':'/content/BBCA.JK_10_years.csv',
    'bbni':'/content/BBNI.JK_10_years.csv',
    'bbri':'/content/BBRI.JK_10_years.csv',
    'bmri':'/content/BMRI.JK_10_years.csv'
}
```

#### langkah selanjutnya adalah dengan kita melakukan looping untuk menjadikan penampungan di atas menjadi satu dataframe

```python
# 2. Kita siapkan keranjang kosong (List) untuk menampung data sementara
list_dataframe = []

# 3. Kita lakukan perulangan (looping) untuk membaca dan memproses file satu per satu
for kode_saham, nama_file in file_saham.items():

    # Membaca file CSV
    df_sementara = pd.read_csv(nama_file)

    # PENTING: Kita tambahkan kolom baru agar saat digabung, kita tahu ini data bank apa
    df_sementara['Kode_Saham'] = kode_saham

    # Masukkan data yang sudah dilabeli ini ke dalam keranjang
    list_dataframe.append(df_sementara)
```

##### Sekarang tinggal kita gabungkan

```python
df = pd.concat(list_dataframe, ignore_index=True)
df.head(),df.tail()
```

#### Sekarang kita akan masuk ke dalam pembersihan data termasuk outlier

```python
df.describe()
```

##### hasil pengamatan
